리뷰의 포인트의 평균을 구하고, 리뷰의 포인트값이, 평균보다 큰 데이터 (즉, 평가가 좋은 와인) 만 가져오시오.

```python
r_mean = reviews.points.mean()
```

```python
reviews.points > r_mean # 이 조건을 만족하는 행을 가져오려면
```

```python
reviews.loc[reviews.points > r_mean , ]
```
